## 감성 사전을 이용해서 각 뉴스기사당 감성 점수 부여하기

```python
knu = pd.read_csv("knu_sentiment_lexicon.csv")

check_total = []
box = []
for i in tqdm(df['TEXT']):
    score = 0
    check = [0,0]
    for w, p in zip(knu['word'], knu['polarity']):
        if w in str(i):
            score += p
            if p > 0:
                check[0] += 1
            elif p < 0:
                check[1] += 1
    check_total.append(check)
    box.append(score)
```

```python
df['SCORE'] = box
df
```

## 감성 점수가 0이상의 정수는 긍정, 아니면 부정 부여하기

```python
box = []
for i in tqdm(df['SCORE']):
    if i >= 0:
        box.append("긍정")
    else:
        box.append("부정")
df['긍정/부정'] = box
df
```

## 긍정, 부정 단어 개수 세기

```python
sample = pd.DataFrame(check_total, columns = ['긍정개수', '부정개수'])
df = pd.concat([df, sample], axis = 1)
df
```

```python
import numpy as np
```

```python
total = []
for i in df.groupby("종목"):
    box = [i[0]]
    sample = df[df['종목'] == i[0]]
    p1 = sample[sample['MONTH'] == '07']["PRICE"].iloc[0]
    p2 = sample[sample['MONTH'] == '08']["PRICE"].iloc[0]
    p3 = sample[sample['MONTH'] == '09']["PRICE"].iloc[0]
    p4 = sample[sample['MONTH'] == '10']["PRICE"].iloc[0]
    p5 = sample[sample['MONTH'] == '11']["PRICE"].iloc[0]
    if p1 < p2:
        box.append(1)
    else:
        box.append(0)
    if p2 < p3:
        box.append(1)
    else:
        box.append(0)
    if p3 < p4:
        box.append(1)
    else:
        box.append(0)
    if p4 < p5:
        box.append(1)
    else:
        box.append(0)
    box.append(None)
    total.append(box)
```

## TARGET열 만들기(다음달보다 종가가 올랐을 때는 1, 아니면 0)

```python
box = []
for i in tqdm(total):
    for j in range(len(df)):
        sample = df.iloc[j]
        if sample['종목'] == i[0] and sample['MONTH'] == '07':
            box.append(i[1])
        elif sample['종목'] == i[0] and sample['MONTH'] == '08':
            box.append(i[2])
        elif sample['종목'] == i[0] and sample['MONTH'] == '09':
            box.append(i[3])
        elif sample['종목'] == i[0] and sample['MONTH'] == '10':
            box.append(i[4])
        elif sample['종목'] == i[0] and sample['MONTH'] == '11':
            box.append(i[5])
df['TARGET'] = box
df
```

## 7월부터 10월 데이터 가져오기

```python
df2 = df[(df['MONTH'] == '07') | (df['MONTH'] == '08') | (df['MONTH'] == '09') | (df['MONTH'] == '10')]
df2.dropna(inplace=True)
df2
```

## 각 종목 합산하기

```python
total = []
for i in tqdm(df2.groupby(["종목", "MONTH"])):
    sample = i[1]
    dsum = sample['SCORE'].sum()
    text = ""
    for j in sample["TEXT"]:
        text += j
        text += " "
    if dsum > 0:
        senti = "긍정"
    else:
        senti = "부정"
    price = sample['PRICE'].iloc[0]
    target = int(sample['TARGET'].iloc[0])
    total.append([text, i[0][0], i[0][1], price, dsum, senti, target])
```

```python
df3 = pd.DataFrame(total, columns = ['TEXT', '종목', 'MONTH', "PRICE",'SCORE', "긍정/부정", "TARGET"])
df3
```

## 감성 사전 점수 정렬하기

```python
df3.sort_values(by = "SCORE", ascending=False, inplace=True)
df3
```

## 랜덤포레스트 모델을 통한 학습

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
news_dtm = cv.fit_transform(df3['TEXT'])
news_dtm
```

```python
data = news_dtm.toarray()
target = df3['TARGET']
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(oob_score=True)
rf.fit(data, target)
```
